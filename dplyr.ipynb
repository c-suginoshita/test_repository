# 関数の連結

```python
import pandas as pd
```

```python
# サンプルデータ
data = dict(x=range(1, 101), y=range(201, 301), z=range(101, 201))
temp1 = pd.DataFrame(data=data, index=range(1, 101))
print(temp1)
```

```python
sub1 = temp1[temp1['x'] > 40]
sub1
```

```python
# x > 50のyの平均(y_mean)とzの中央値(z_median)のデータを作る
sub = temp1[temp1['x'] > 50]
y_mean = sub.y.mean()
z_median = sub.z.median()

#dat1 <- data.frame(y_mean = y_mean, z_median = z_median)
dat1 = pd.DataFrame(data={'y_mean':[y_mean], 'z_median':[z_median]})
dat1
```

```python
# 使用の例
dat2 = pd.DataFrame(data={'y_mean':[temp1.query('x > 50').y.mean()], 'z_median':[temp1.query('x > 50').z.median()]})
dat2
```
